### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
import pandas as pd

# 1. Contando as vendas por categoria em cada loja
cat1 = loja['Categoria do Produto'].value_counts()
cat2 = loja2['Categoria do Produto'].value_counts()
cat3 = loja3['Categoria do Produto'].value_counts()
cat4 = loja4['Categoria do Produto'].value_counts()

# 2. Criando a tabela consolidada (unindo as 4 lojas)
tabela_vendas_cat = pd.concat([cat1, cat2, cat3, cat4], axis=1)
tabela_vendas_cat.columns = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# 3. Adicionando a linha de TOTAL no final
# O .sum() vai somar todas as vendas de cada coluna (loja)
tabela_vendas_cat.loc['TOTAL DE VENDAS 🚀'] = tabela_vendas_cat.sum()

# 4. Tratando os dados (preenchendo vazios com 0 e deixando números inteiros)
tabela_final = tabela_vendas_cat.fillna(0).astype(int)

# Exibindo o resultado 🤩💯
print("--- PERFORMANCE POR CATEGORIA E VOLUME TOTAL ---")
print(tabela_final)
```

# 3. Média de Avaliação das Lojas


```python
# Calculando a média de avaliação por loja


media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

# Mostrando os resultados formatados
print(f"Média de Avaliação - Loja 1: {media_loja1:.2f} ⭐")
print(f"Média de Avaliação - Loja 2: {media_loja2:.2f} ⭐")
print(f"Média de Avaliação - Loja 3: {media_loja3:.2f} ⭐")
print(f"Média de Avaliação - Loja 4: {media_loja4:.2f} ⭐")
```

```python
# Calculando a satisfação por categoria em cada loja

media_cat_loja1 = loja.groupby('Categoria do Produto')['Avaliação da compra'].mean()
media_cat_loja2 = loja2.groupby('Categoria do Produto')['Avaliação da compra'].mean()
media_cat_loja3 = loja3.groupby('Categoria do Produto')['Avaliação da compra'].mean()
media_cat_loja4 = loja4.groupby('Categoria do Produto')['Avaliação da compra'].mean()

# Mostrando o resultado da Loja 1 para analisar
print("Média de Avaliação por Categoria - Loja 1:")
print(media_cat_loja1)
```

```python
import pandas as pd

# 1. Calculando as médias por categoria para cada loja
media_cat1 = loja.groupby('Categoria do Produto')['Avaliação da compra'].mean()
media_cat2 = loja2.groupby('Categoria do Produto')['Avaliação da compra'].mean()
media_cat3 = loja3.groupby('Categoria do Produto')['Avaliação da compra'].mean()
media_cat4 = loja4.groupby('Categoria do Produto')['Avaliação da compra'].mean()

# 2. Juntando tudo em uma única tabela (DataFrame)
tabela_comparativa = pd.concat([media_cat1, media_cat2, media_cat3, media_cat4], axis=1)

# 3. Nomeando as colunas para ficar profissional 💅
tabela_comparativa.columns = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# 4. Formatando com 2 casas decimais, trocando ponto por vírgula e colocando a estrela ⭐
# Usamos o applymap para aplicar a regra em todas as células da tabela
tabela_final = tabela_comparativa.applymap(lambda x: f"{x:.2f} ⭐".replace('.', ','))

# Exibindo o resultado final 🤩💯
print("--- MÉDIA DE AVALIAÇÃO POR CATEGORIA ---")
print(tabela_final)
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd

# 1. Criando um dicionário para organizar os destaques
destaques_vendas = {}

# Lista com os seus DataFrames (loja, loja2, loja3 e loja4)
minhas_lojas = [loja, loja2, loja3, loja4]

for i, df in enumerate(minhas_lojas, 1):
    # Contamos as vendas de cada produto
    contagem = df['Produto'].value_counts()

    # Extraímos o nome e a quantidade do MAIS e do MENOS vendido
    # idxmax/idxmin pegam o nome do produto; max/min pegam o número de vendas
    destaques_vendas[f'Loja {i}'] = [
        contagem.idxmax(),             # Nome do Mais Vendido
        f"{contagem.max()} un.",       # Qtd do Mais Vendido
        contagem.idxmin(),             # Nome do Menos Vendido
        f"{contagem.min()} un."        # Qtd do Menos Vendido
    ]

# 2. Criando a tabela final com as linhas descritivas
tabela_vendas = pd.DataFrame(destaques_vendas, index=[
    'Produto Mais Vendido 🏆',
    'Quantidade (Máx) 📈',
    'Produto Menos Vendido 📉',
    'Quantidade (Mín) 🛒'
])

# Exibindo o resultado final 💅💯
print("--- RELATÓRIO DE PERFORMANCE DE PRODUTOS ---")
print(tabela_vendas)
```

# 5 Frete Médio por Loja

```python
import pandas as pd

# 1. Calculando a média de frete para cada uma das 4 lojas
media_f1 = loja['Frete'].mean()
media_f2 = loja2['Frete'].mean()
media_f3 = loja3['Frete'].mean()
media_f4 = loja4['Frete'].mean()

# 2. Criando a tabela consolidada 📋✨
resumo_frete = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Custo Médio de Frete': [media_f1, media_f2, media_f3, media_f4]
})

# 3. Formatando para o padrão brasileiro (R$, vírgula e 2 casas)
resumo_frete['Custo Médio de Frete'] = resumo_frete['Custo Médio de Frete'].apply(
    lambda x: f"R$ {x:.2f}".replace('.', ',')
)

# Exibindo o resultado final
print("--- RELATÓRIO DE LOGÍSTICA (FRETE) ---")
print(resumo_frete.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Preparando os dados para o gráfico
lojas_lista = [loja, loja2, loja3, loja4]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

mais_vendidos_nomes = []
mais_vendidos_qtd = []
menos_vendidos_nomes = []
menos_vendidos_qtd = []

# Coletando os extremos de cada loja
for df in lojas_lista:
    contagem = df['Categoria do Produto'].value_counts()
    # Pegando o Top 1 🏆
    mais_vendidos_nomes.append(contagem.idxmax())
    mais_vendidos_qtd.append(contagem.max())
    # Pegando o Último Lugar 📉
    menos_vendidos_nomes.append(contagem.idxmin())
    menos_vendidos_qtd.append(contagem.min())

# 2. Configurando o gráfico de barras duplas
x = np.arange(len(nomes_lojas)) # Posição das lojas
largura = 0.40 # Largura das barras

fig, ax = plt.subplots(figsize=(14, 8))

# Criando as barras
barra1 = ax.bar(x - largura/2, mais_vendidos_qtd, largura, label='Mais Vendida 👑', color='#FF1493') # Rosa Choque
barra2 = ax.bar(x + largura/2, menos_vendidos_qtd, largura, label='Menos Vendida 🧊', color='#00BFFF') # Azul Turquesa

# 3. Estilização e Storytelling 👄✨
ax.set_ylabel('Quantidade de Vendas', fontsize=12, fontweight='bold')
ax.set_title('Categorias: O que mais vende vs. O que menos vende', fontsize=16, fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(nomes_lojas, fontsize=12, fontweight='bold')
ax.legend(fontsize=12)

# Adicionando os nomes das categorias em cima das barras para ninguém ter dúvida!
def colocar_labels(barras, categorias, valores):
    for i, barra in enumerate(barras):
        altura = barra.get_height()
        ax.annotate(f'{categorias[i]}\n({valores[i]})',
                    xy=(barra.get_x() + barra.get_width() / 2, altura),
                    xytext=(0, 5), # 5 pontos de distância vertical
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=10, fontweight='bold')

colocar_labels(barra1, mais_vendidos_nomes, mais_vendidos_qtd)
colocar_labels(barra2, menos_vendidos_nomes, menos_vendidos_qtd)

plt.ylim(0, max(mais_vendidos_qtd) + 100) # Espaço extra pro texto
plt.tight_layout()

# Salvando a obra de arte 🖼️💯
plt.savefig('performance_categorias_lojas.png')
```
